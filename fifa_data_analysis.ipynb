# **Analysis**

## Q) Which teams scored the most total goals in the tournament?

```python
df.groupby("Team")['Goal'].sum().sort_values(ascending=False).head(10).plot(kind='bar')
plt.title("Top 10 teams that scored the most goals")
```

## Q) Which teams conceded the most goals?

```python
df.groupby("Against")['Goal'].sum().sort_values(ascending=False).head(10).plot(kind='bar')
plt.title("Top 10 teams that conceded the most goals")
```

Q) Which match had the highest total goals?

```python
df.groupby("Match No.")["Goal"].sum().sort_values(ascending=False).head(1)
```

##Q) Find out how many times the teams are played in this Fifa Worldcup-2022. On top of this, find out the ranks of the teams.

```python
teams_played = df["Team"].value_counts().reset_index()
teams_played.columns = ["Team", "Counts"]
teams_played['rank'] = teams_played["Counts"].rank(method='first', ascending=False)
teams_played.set_index("Team", inplace=True)
teams_played
```

##Q) Do these below operations:

Find out the rank based on the "Team" column and save the result by adding a new column named "Rank".

Change the datatype of this column to integer by using np.int16
Set the index of the DataFrame by using this "Rank" column permanently.

After that, sort the dataframe based on the "Rank" index.

```python
rank = df["Team"].value_counts().rank(method="first", ascending=False).astype('int16')
df['Rank'] = df["Team"].map(rank)
df.set_index("Rank",inplace=True)
df.sort_index()
```

##Q) Which teams appeared most frequently in matches?

```python
df[['Team','Against']].stack().value_counts().head(5).plot(kind='bar')
plt.title("Top 5 teams with Most Frequent Matches")
```
